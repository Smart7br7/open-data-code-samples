```javascript
const fetch=require('node-fetch');
const d3=require('d3');
```

```javascript
const API_ENDPOINT='https://opendataprod.br7.org.il/api/3/action'
```

```javascript
const fetchOptions = {
    redirect: 'follow',
};
```

## Getting the list of avaulable packages

```javascript
$$.async();
fetch(`${API_ENDPOINT}/package_list`, fetchOptions)
    .then((result) => result.json())
    .then((result) => result.result)
    .then((package_list) => {
        console.log(package_list);
        $$.done();
    });
```

## Getting details about a single package

```javascript
$$.async();
fetch(`${API_ENDPOINT}/package_show?id=dog-gardens`, fetchOptions)
    .then((result) => result.json())
    .then((result) => result.result)
    .then((package_info) => {
        console.log(package_info);
        $$.done();
    });
```

## Processing data from a single resource

```javascript
$$.async();
fetch(`${API_ENDPOINT}/package_show?id=dog-gardens`, fetchOptions)
    .then((result) => result.json())
    .then((result) => result.result)
    .then((package_info) => package_info.resources
            .filter((r) => r.format === 'GeoJSON')[0].url)
    .then((geojson_url) => fetch(geojson_url, fetchOptions))
    .then((result) => result.json())
    .then((geojson) => {
        console.log('This is GeoJSON:');
        console.log(geojson);
        $$.done();
    });
```

```javascript
$$.async();
fetch(`${API_ENDPOINT}/package_show?id=tender`, fetchOptions)
    .then((result) => result.json())
    .then((result) => result.result)
    .then((package_info) => package_info.resources
            .filter((r) => r.format === 'JSON')[0].url)
    .then((json_url) => fetch(json_url, fetchOptions))
    .then((result) => result.json())
    .then((json) => {
        console.log('This is JSON:');
        console.log(json);
        $$.done();
    });
```
