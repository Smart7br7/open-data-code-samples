```python
import requests
```

```python
API_ENDPOINT='https://opendataprod.br7.org.il/api/3/action'
```

## Getting the list of available packages

```python
package_list = requests.get(f'{API_ENDPOINT}/package_list').json()['result']
package_list
```

## Getting details about a single package

```python
package_info = requests.get(f'{API_ENDPOINT}/package_show?id=dog-gardens').json()['result']
package_info
```

## Processing data from a single resource

```python
package_info = requests.get(f'{API_ENDPOINT}/package_show?id=dog-gardens').json()['result']
geojson_url = next(filter(lambda r: r['format'] == 'GeoJSON', package_info['resources']))['url']
geojson = requests.get(geojson_url).json()
print('This is GEOJson')
geojson
```

```python
package_info = requests.get(f'{API_ENDPOINT}/package_show?id=tender').json()['result']
json_url = next(filter(lambda r: r['format'] == 'JSON', package_info['resources']))['url']
content = requests.get(json_url).json()
print('This is JSON')
content
```
